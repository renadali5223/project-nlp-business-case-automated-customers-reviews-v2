```python
import pandas as pd

df = pd.read_csv('/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')

df.head()
```

```python
df.columns
```

```python
df = df[['reviews.rating', 'reviews.text']]

df.dropna(subset=['reviews.rating', 'reviews.text'], inplace=True)

df.head()
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'<.*?>', '', text)  # remove HTML tags
    text = re.sub(r'http\S+', '', text)  # remove URLs
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra whitespace
    return text

df['cleaned_review'] = df['reviews.text'].apply(clean_text)

df[['reviews.text', 'cleaned_review']].head()
```

```python
def get_sentiment(rating):
    if rating in [1, 2]:
        return 'Negative'
    elif rating == 3:
        return 'Neutral'
    elif rating in [4, 5]:
        return 'Positive'

df['sentiment'] = df['reviews.rating'].apply(get_sentiment)

df[['reviews.rating', 'sentiment']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sentiment', order=['Negative', 'Neutral', 'Positive'], palette='Set2')
plt.title('Distribution of Sentiment Classes')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
df['reviews.text'] = df['reviews.text'].str.replace(r'\W', ' ', regex=True)

df.head()
```

```python
import numpy as np

# تصنيف المراجعات بناءً على التقييم
def map_rating_to_sentiment(rating):
    if rating <= 2:
        return "negative"
    elif rating == 3:
        return "neutral"
    else:
        return "positive"

df["sentiment"] = df["reviews.rating"].apply(map_rating_to_sentiment)

# طباعة توزيع الفئات
print(df["sentiment"].value_counts())
```

```python
from sklearn.model_selection import train_test_split

# تقسيم البيانات
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df["reviews.text"].tolist(), df["sentiment"].tolist(), test_size=0.2, random_state=42)

# طباعة عدد العينات
print(f"Train size: {len(train_texts)}")
print(f"Test size: {len(test_texts)}")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_labels_enc = le.fit_transform(train_labels)
test_labels_enc = le.transform(test_labels)

# حفظ أسماء الفئات
label_names = le.classes_.tolist()
print("Labels:", label_names)
```

```python
from transformers import DistilBertTokenizerFast

tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

# ترميز النصوص
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=512)
test_encodings = tokenizer(test_texts, truncation=True, padding=True, max_length=512)
```

```python
import torch

class ReviewDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item
    def __len__(self):
        return len(self.labels)

train_dataset = ReviewDataset(train_encodings, train_labels_enc)
test_dataset = ReviewDataset(test_encodings, test_labels_enc)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

from transformers import TrainingArguments
```

```python
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments

model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased", num_labels=3)

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    logging_dir="./logs",
    logging_steps=10,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
```

```python
trainer.train()
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report
import numpy as np

# استخدم trainer إذا كنت دربت به
predictions = trainer.predict(test_dataset)

# نحصل على القيم المتوقعة
y_pred = np.argmax(predictions.predictions, axis=1)

# القيم الحقيقية
y_true = predictions.label_ids
```

```python
# الدقة (Accuracy)
accuracy = accuracy_score(y_true, y_pred)

# Precision, Recall, F1 لكل فئة
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average=None)

# المصفوفة التصادمية (Confusion Matrix)
conf_matrix = confusion_matrix(y_true, y_pred)

# تقرير شامل
report = classification_report(y_true, y_pred, target_names=["Negative", "Neutral", "Positive"])
```

```python
print("Accuracy:", round(accuracy * 100, 2), "%")
print("\nClassification Report:\n", report)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
tokenizer_t5 = T5Tokenizer.from_pretrained("t5-small")
model_t5 = T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
def summarize_reviews_t5(text, max_input_length=512, max_output_length=150):
    # تحضير الإدخال
    input_text = "summarize: " + text
    inputs = tokenizer_t5.encode(input_text, return_tensors="pt", max_length=max_input_length, truncation=True)

    # التوليد
    outputs = model_t5.generate(inputs, max_length=max_output_length, num_beams=4, early_stopping=True)

    # استخراج النص
    return tokenizer_t5.decode(outputs[0], skip_special_tokens=True)
```

```python
summary_per_cluster = {}

for cluster_id in range(num_clusters):
    print(f"\nGenerating summary for Cluster {cluster_id}...\n")

    # استخراج المراجعات من هذا الكلاستر
    cluster_reviews = df[df['cluster'] == cluster_id]['cleaned_review'].tolist()

    # جمع أول 100 مراجعة
    combined_reviews = " ".join(cluster_reviews[:100])

    # تلخيص
    summary = summarize_reviews_t5(combined_reviews)

    summary_per_cluster[cluster_id] = summary
    print(summary)
```

```python
import gradio as gr

# 🔹 Import required libraries
import tensorflow as tf
import numpy as np

# Load models and tokenizers
from transformers import DistilBertTokenizerFast, TFDistilBertForSequenceClassification
from transformers import T5Tokenizer, T5ForConditionalGeneration
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd
import re

# 🔹 Load dataset
df = pd.read_csv('Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
df = df[['name', 'reviews.text', 'primaryCategories']].dropna()

# Text cleaning function
def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_review'] = df['reviews.text'].apply(clean_text)
df['full_text'] = df['name'] + " " + df['cleaned_review']

# 🔹 Feature extraction and KMeans clustering
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(df['full_text'])

kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# 🔹 Load sentiment classification model
tokenizer_cls = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
model_cls = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)

# 🔹 Load T5 summarization model (not used in real-time here)
tokenizer_t5 = T5Tokenizer.from_pretrained("t5-small")
model_t5 = T5ForConditionalGeneration.from_pretrained("t5-small")

# 🔹 Predefined summaries for each cluster
summary_per_cluster = {
    0: "Recommendation: This cluster contains positive reviews for e-readers like Kindle, with minor complaints about ads.",
    1: "Recommendation: Focuses on mobile accessories. The best ones are durable; the worst have charging issues.",
    2: "Recommendation: Contains negative reviews about cheap headphones. Best to avoid them.",
    3: "Recommendation: Mixed reviews for smart devices like Alexa. Mostly positive, with privacy concerns.",
    4: "Recommendation: General laptop reviews. Best are MacBooks; worst come from unknown brands."
}

# 🔹 Main analysis function
def analyze_review(review_text):
    cleaned = clean_text(review_text)

    # Sentiment classification
    tokens = tokenizer_cls(cleaned, return_tensors='tf', truncation=True, padding=True)
    logits = model_cls(**tokens).logits
    pred_class = tf.argmax(logits, axis=1).numpy()[0]
    sentiment = ["Positive", "Neutral", "Negative"][pred_class]

    # Cluster prediction
    full_input = "product " + cleaned
    tfidf = vectorizer.transform([full_input])
    cluster_id = kmeans.predict(tfidf)[0]

    # Get recommendation summary for the cluster
    summary = summary_per_cluster[cluster_id]

    return sentiment, f"Cluster #{cluster_id}", summary

# 🔹 Create Gradio interface
interface = gr.Interface(
    fn=analyze_review,
    inputs=gr.Textbox(label="Enter your review here"),
    outputs=[
        gr.Label(label="Sentiment Classification"),
        gr.Label(label="Cluster Number"),
        gr.Textbox(label="Recommendation Summary")
    ],
    title="Amazon Reviews Analysis",
    description="Enter an Amazon product review, and this AI tool will analyze it by classifying sentiment, identifying the product group, and showing a tailored recommendation."
)

interface.launch()
```
